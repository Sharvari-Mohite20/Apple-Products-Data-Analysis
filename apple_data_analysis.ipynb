#### Project on Apple Products Data Analysis

#### Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load Dataset

```python
df = pd.read_csv("apple_products.csv")
df.head()
```

#### Dataset Overview

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

#### Check Missing Values

```python
df.isnull().sum()
```

#### Check Duplicate Records

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

#### Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

#### Data Cleaning

```python
df["Ram"] = df["Ram"].str.replace(" GB", "")

df["Ram"] = pd.to_numeric(df["Ram"])

df.head()
```

#### Check Datatypes

```python
df.dtypes
```

#### Basic Statistics

```python
df[[
    "Sale Price",
    "Mrp",
    "Discount Percentage",
    "Star Rating"
]].describe()
```

#### Top 10 Most Expensive Products

```python
top_expensive = df.sort_values(
    by="Sale Price",
    ascending=False
)

top_expensive[
    ["Product Name","Sale Price"]
].head(10)
```

#### Top Rated Products

```python
top_rated = df.sort_values(
    by="Star Rating",
    ascending=False
)

top_rated[
    ["Product Name","Star Rating"]
].head(10)
```

#### Price Distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["Sale Price"],
    bins=15,
    kde=True
)

plt.title("Distribution of Apple Product Prices")
plt.xlabel("Sale Price")
plt.ylabel("Count")

plt.savefig("images/distribution_appleprod_prices.png")

plt.show()
```

#### Top 10 Products by Ratings

```python
top_ratings = df.nlargest(
    10,
    "Number Of Ratings"
)

top_ratings
```

#### Visualization

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x="Number Of Ratings",
    y="Product Name",
    data=top_ratings
)

plt.title("Top 10 Products by Number of Ratings")

plt.savefig("images/top10_prod_by_ratings.png")

plt.show()
```

#### Top 10 Products by Reviews

```python
top_reviews = df.nlargest(
    10,
    "Number Of Reviews"
)

plt.figure(figsize=(12,6))

sns.barplot(
    x="Number Of Reviews",
    y="Product Name",
    data=top_reviews
)

plt.title("Top 10 Products by Reviews")

plt.savefig("images/top10_prod_by_reviews.png")

plt.show()
```

#### Average Sale Price by RAM

```python
ram_price = df.groupby("Ram")[
    "Sale Price"
].mean()

ram_price
```

#### RAM vs Average Price

```python
plt.figure(figsize=(8,5))

ram_price.plot(
    kind="bar"
)

plt.title("Average Sale Price by RAM")
plt.ylabel("Average Price")

plt.savefig("images/avg_sales_price_by_ram.png")

plt.show()
```

#### Discount Distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["Discount Percentage"],
    bins=8
)

plt.title("Discount Percentage Distribution")

plt.savefig("images/discount_percentage_distribution.png")

plt.show()
```

#### Correlation Analysis

```python
corr = df[
    [
        "Sale Price",
        "Mrp",
        "Discount Percentage",
        "Number Of Ratings",
        "Number Of Reviews",
        "Star Rating",
        "Ram"
    ]
].corr()

corr
```

#### Correlation Heatmap

heatmap displays the correlation between these variables:

Sale Price
MRP
Discount Percentage
Number Of Ratings
Number Of Reviews
Star Rating
RAM

+1 → Perfect positive relationship
0 → No relationship
-1 → Perfect negative relationship

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("images/correlation_heatmap.png")

plt.show()
```

#### Sale Price vs Ratings

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Sale Price",
    y="Number Of Ratings",
    data=df
)

plt.title("Price vs Ratings")

plt.savefig("images/price_vs_ratings.png")

plt.show()
```

#### Highest Discount Products

```python
highest_discount = df.sort_values(
    by="Discount Percentage",
    ascending=False
)

highest_discount[
    [
        "Product Name",
        "Sale Price",
        "Discount Percentage"
    ]
].head(10)
```

#### Star Rating Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Star Rating",
    data=df
)

plt.title("Distribution of Star Ratings")

plt.savefig("images/dist_of_star_rating.png")


plt.show()
```

#### Key Insights

1. Apple products in the dataset are concentrated in the premium price range, reflecting the brand's premium market positioning.

2. Sale Price and MRP exhibit a very strong positive correlation, indicating that discounted prices closely follow the original pricing structure.

3. A small number of products account for a large share of ratings and reviews, indicating that customer engagement is concentrated among a few popular products.

4. Products with more ratings generally receive more reviews, showing a strong relationship between product popularity and customer feedback.

5. Products with larger RAM configurations tend to have higher average selling prices, demonstrating the impact of hardware specifications on pricing.

6. Most products maintain star ratings above 4.0, indicating generally positive customer experiences.

7. Several high-priced Apple products continue to attract substantial numbers of ratings and reviews, indicating strong customer interest despite premium pricing.

8. Discount percentages vary significantly across products, suggesting selective promotional strategies rather than uniform discounting.

9. Many of the most-rated products also maintain strong star ratings, indicating a positive relationship between popularity and customer satisfaction.

10. RAM and Sale Price show a positive correlation, suggesting that storage capacity contributes to pricing, although additional factors also influence product cost.
